# Introduction

Air pollution is a public health and environmental issue (Pope and Dockery, 2006). The health effects of poor air quality can cause serious health problems, such as increased morbidity and mortality in the respiratory, cardiovascular and cerebrovascular systems (WHO, 2013). The impact of air pollution also differs by socioeconomic status for individuals and in developing countries, hence considering it to be another factor towards the contribution of global health inequality. In addition to impact on health, air pollution leads to greater economic costs, such as impact on agricultural crops and damage to buildings and infrastructure (WHO, 2021). WHO (2021) states that air pollution-related climate changes are also associated with higher costs and environmental degradation. They also state that air pollution results in the loss of trillions of dollars, loss of healthy life, and cause millions of deaths annually.

As it has been associated with asthma, lung cancer, and lower life expectancy (Wang, 2016), an essential tool for informing the public about air quality is the Air Quality Index (AQI), that summarizes the concentrations of different pollutants such as particulate matter, ozone, sulfur dioxide, carbon monoxide, and nitrogen dioxide (NO<sub>2</sub>). Its importance resides in the fact that it offers a clear, easily understandable metric that helps people, groups, and authorities recognise and manage the possible health risks linked to air pollution. 

Nitrogen dioxide (NO<sub>2</sub>)  linked to industrial processes and vehicle combustion is a particularly notable indicator among the pollutants monitored by the Air Quality Index (AQI). 
Despite efforts to improve air quality, many countries are still impacted by the effects of air pollution (Li, 2016). The respiratory irritant NO<sub>2</sub> is a major factor in determining the general quality of the air. Hence, it is crucial to comprehend NO<sub>2</sub> levels in various regions for a number of reasons. In the first place, it makes it possible to pinpoint locations with high pollution levels and implement focused interventions to enhance air quality. It also functions as a tool for assessing the efficacy of programmes and policies aimed at controlling pollution. Furthermore, the comparative study of NO<sub>2</sub> levels aids in the development of strategies to lower emissions and improve air quality by offering insights into the various sources of pollution.
 
Hence, the aim of this data science project is to analyze the effects of NO<sub>2</sub>, a measure of air quality, in different regions, countries, and between different years within countries. The analysis begins with comparing the WHO classified regions, the European Region and the South East Asia Region. A country was also analyzed within each region. These countries are France and India. In these countries, the difference in NO<sub>2</sub> was analyzed between two different years within the same country. Finally, to understand if global NO<sub>2</sub> has been changing with time, NO<sub>2</sub> emissions were classified and tested for independence between years.

## Methodology

### T test

When two independent groups have different variances and unequal sample sizes, a statistical test called the Welch t-test, also called the Welch's unequal variances t-test, is used to compare the means of the two groups. It is a development of the conventional Student's t-test, which counts on the two groups' variances being equal.

Usually the hypothesis is formulated for this, where 

$H_0$= The means of the output are equal 

$H_1$= The means of the output are not equal (You can specify 'greater' than or 'less' than too)


The test statistic for the Welch t-test is calculated as:

$$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

where $(\bar{X}_1, \bar{X}_2)$ are the sample means, $(s_1^2, s_2^2)$ are the sample variances, and $(n_1, n_2)$ are the sample sizes of the two groups.

The degrees of freedom for the Welch t-test are calculated using the Welch-Satterthwaite equation:
$$
\text{df} \approx \frac{\left(\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}\right)^2}{\frac{\left(\frac{s_1^2}{n_1}\right)^2}{n_1 - 1} + \frac{\left(\frac{s_2^2}{n_2}\right)^2}{n_2 - 1}}
$$

The test statistic calculated is then calculated on basis of which one either rejects or accepts the null hypothesis. 

The Welch t-test makes the assumption that data is measured on an interval or ratio scale, and that the populations from which samples are taken have a roughly normal distribution. It is also assumed that random sampling, independence of observations, and homogeneity of variances within groups exist. Although the test is resistant to deviations from normalcy and unequal variances, the outcomes could be affected by extreme deviations. When comparing the means of independent groups with unequal variances, it works well. When variances between groups are not equal, the Welch t-test performs better than the Student's t-test. Without assuming equal variances, it yields accurate results, especially in small sample sizes. Because of this, it is a strong option for comparing means in situations where the Student's t-test may not be as trustworthy. (Duke University, n.d.; Zach, 2019; Glen, n.d.)


### Chi Square Test

The chi-square test of independence evaluates if two categorical variables significantly correlate with one another. In a contingency table, the observed and expected frequencies are compared as part of the test. As part of the test, a contingency table showing the observed frequencies of the variables' joint occurrences must be created. On the presumption that the variables are independent, expected frequencies are computed. Next, the chi-square statistic is calculated to determine how much the observed and expected frequencies differ from one another. A greater departure from independence is indicated by a higher chi-square value.

To calculate the chi-square test of independance, you set up the null and the alternative hypothesis first where 

$H_0$ : There is absence of significant association between the two variables 

$H_1$ : There is presence of significant association between the two variables

The contingency table looks like this:

$$
\begin{array}{cccccc}
    & \text{Category 1} & \text{Category 2} & \ldots & \text{Category j} & \text{Total} \\
\text{Group A} & O_{11} & O_{12} & \ldots & O_{1j} & R_1 \\
\text{Group B} & O_{21} & O_{22} & \ldots & O_{2j} & R_2 \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
\text{Group i} & O_{i1} & O_{i2} & \ldots & O_{ij} & R_i \\
\text{Total} & C_1 & C_2 & \ldots & C_j & N \\
\end{array}
$$


The expected frequencies are calculated using;

$$
E_{ij} = \frac{\text{Sum of row } i \times \text{Sum of column } j}{\text{Grand total}}
$$

The chi-square is calculated using:

$$
\chi^2 = \sum_{i=1}^{r} \sum_{j=1}^{c} \frac{(O_{ij} - E_{ij})^2}{E_{ij}}
$$

where $r$ is the number of rows and $c$ is the number of columns

The degrees of freedom is calculated using: 

$$
df = (r-1) \cdot (c-1)
$$

The null hypothesis is rejected or accepted if the calculated chi-square statistic is less than 0.05 after comparing it to the critical value from the chi-square distribution table with df degrees of freedom. 

The chi-square test of independence relies on the assumptions of mutually exclusive categories, random sampling, and categorical data. For results that can be trusted, a sufficient sample size is needed. To ensure the validity of the test results, the test also assumes the independence of observations and that the expected frequency in each cell is not too low. (Statistics Solutions, n.d.; Penn State World Campus, n.d.; University of Utah, n.d.)

### Bootstrapping

A statistical technique called bootstrapping generates numerous simulated samples by resampling a single dataset. With a variety of sample statistics, this procedure enables one to compute standard errors, creating confidence intervals, and carrying out the hypothesis testing. Known for being more widely applicable and easier to comprehend than traditional hypothesis testing techniques, bootstrap methods are an alternative. This approach is especially useful when sample sizes are constrained or when it is difficult to determine the underlying statistical properties analytically. 

In order to bootstrap a distribution, multiple datasets of the same size as the original sample are created, a selected statistic of interest is determined for each dataset, and a distribution is then built using the resampled statistics. By acting as a stand-in for the statistic's sampling variability, this distribution makes it easier to estimate confidence intervals and conduct hypothesis testing. Next, bias can be computed from both the boostrapped sample and the original sample to determine which one performs better, based on the user-defined question. 

Compared to parametric methods, boostrapping has the advantage of requiring fewer distributional assumptions; however, it still relies on the assumption that the observed data accurately represents the underlying population distribution. Additionally, it relies on the randomness of the sampling process and the independence of observations among samples to produce accurate and impartial results. (Hickman, 2023; University of Washington, n.d.)

## Test 1: Did South East Asia Region have more NO<sub>2</sub> (µg/m<sup>3</sup>) emissions than the European Region in 2019?  

To compare the average NO<sub>2</sub> (µg/m<sup>3</sup>) in the South East Asia Region compared to the European Region, a Welch Two Sample t-test is performed. In preparation for t-test, the assumptions of the test are assessed before comparing the appropriate variables. First, each observation is assumed to be independent of each observation in the sample. Each region included NO<sub>2</sub> (µg/m<sup>3</sup>) observations for each city in each respective region in the year 2019. Second, each city is assumed to be randomly sampled in the dataset. Third, in this subset, the number of observations for each region had the best balance compared to other years and other regions. The South East Asia Region and European Region has the greatest number of samples in 2019. Finally, the data is reviewed for normality (Statology, 2021).


```r
library(dplyr)
df <- read.csv('data_final.csv')
df1 <- subset(df, select=c(Measurement.Year, WHO.Country.Name))
df2 <- df1[order(df1$Measurement.Year), ]
table <- table(df1)

# Create a new data frame with counts for each combination of Year and Country
count_df <- df1 %>%
  group_by(Measurement.Year, WHO.Country.Name) %>%
  summarize(Frequency = n())

colnames(df)[colnames(df) == "NO2....g.m3."] <- "NO2"
colnames(df)[colnames(df) == "NO2.temporal.coverage...."] <- "NO2.temporal.coverage"
#print(head(df))

df_region <- subset(df, select=c(Measurement.Year, WHO.Region))
```

An important criterion in t-test is that the compared samples should have similar sample sizes. From the clean dataset, a table is created below to count the number of samples in each region by each year. Additionally, another table is created to review the number of samples in each country by each year. Due to the limitations in the sample sizes of each WHO Region, the regions with the highest frequencies are selected for analysis. However, whilst selecting the sample sizes for the Europen Region and South East Asia Region in terms of NO<sub>2</sub> (µg/m<sup>3</sup>), the sample sizes are shown to be different, due to which Welch T-test is used. 

```r
# Create a new data frame with counts for each combination of Year and Country
df_region <- df_region %>%
  group_by(Measurement.Year, WHO.Region) %>%
  summarize(Frequency = n())
df_region <- df_region[df_region$Measurement.Year==2019,]
# Display the resulting data frame
print(df_region)
```

Consequently, the assumption of normality is explored in depth, wherein it is seen that both the South East Asia Region and the European Region are approximately normally distributed. 

```r
df_SEA_2019 <- df$NO2[df$WHO.Region=="South East Asia Region" & 
                                df$Measurement.Year=="2019"]

qqnorm(df_SEA_2019, main="Normal Q-Q Plot of NO2 in South East Asia Region in 2019")
qqline(df_SEA_2019, col = "red", lwd = 2)
```

```r
df_EUR_2019 <- df$NO2[df$WHO.Region=="European Region" & 
                                df$Measurement.Year=="2019"]

qqnorm(df_SEA_2019, main="Normal Q-Q Plot of NO2 in European Region in 2019")
qqline(df_SEA_2019, col = "blue", lwd = 2)
```

For the statistical testing, the following are null hypothesis and alternative hypothesis using Welch Two Sample t-test


$H_0$: The NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in the South East Asia Region and European Region are similar.
      
$H_1$: The NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in the South East Asia Region is greater than in the European Region.  



```r
df2019 <- df[df$Measurement.Year==2019, ]

sea2 <- subset(df2019, select=NO2,
                      subset=WHO.Region=="South East Asia Region", drop=T)

eur2  <- subset(df2019, select=NO2,
                      subset=WHO.Region=="European Region", drop=T)

t.test(sea2, eur2, alt="greater")
```

Conclusion: p-value was less than 0.05, which rejects the null hypothesis

## Test 2.1: Does France have less NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2016 than in 2010? 

Given that South East Asian has greater NO<sub>2</sub> (µg/m<sup>3</sup>) emissions than Europe, additional testing was performed on France that belongs to the European Region and India from the South East Asian Region

To test this hypothesis, two-sample bootstrap test for the ratio of means of NO<sub>2</sub> (µg/m<sup>3</sup>) is performed. Bootstrap testing is a powerful tool for this dataset. Although the original data has over 32,190 observations, the dataset decreased to 16,364 after excluding columns with majority missing data and low temporal coverage. When choosing the variables to analyze, the variables with the most complete data are selected. In France, this resulted in about 286 observations in 2018 and 304 observations in 2010, which are closer to each other in terms of frequency counts. Hence, the two-sample bootstrap test can be performed on this constrained sample size between the years 2010 and 2016. 

```r
count_df_france <- count_df[count_df$WHO.Country.Name=="France",]
print(count_df_france)
```

Using the two-sample boostrap test for the ratio of means of NO<sub>2</sub> (µg/m<sup>3</sup>), the null hypothesis and alternative hypothesis are:  

$H_0$: France did not decrease NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2016 compared to 2010.  

$H_1$: France did decrease NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2016 compared to 2010.

```r
df_france <- df[df$WHO.Country.Name=="France", ]

france2010 <- subset(df_france, select=NO2,
                      subset=Measurement.Year=="2010", drop=T)

france2016  <- subset(df_france, select=NO2,
                      subset=Measurement.Year=="2016", drop=T)

t.test(france2016, france2010, alt="less")
```

The boostrapping for this is performed with a sample size of 100,000 using the ratio of means. After the calculation of the bias of 0.03 which is almost non-existant, the boostrap test shows the 95% confidence in the alternative hypothesis. 

```r
df_france <- df[df$WHO.Country.Name=="France", ]
df_france_2010 <- df_france[df_france$Measurement.Year == "2010", ]
df_france_2016 <- df_france[df_france$Measurement.Year == "2016", ]
N <- 100000

ratiomeans<- numeric(N)

for (i in 1:N)
{
  fr_sample_2010 <- sample(df_france_2010$NO2[df_france_2010$Measurement.Year == '2010'], length(df_france_2010), replace = TRUE)
  fr_sample_2016 <- sample(df_france_2016$NO2[df_france_2016$Measurement.Year == '2016'], length(df_france_2016), replace = TRUE)
  ratiomeans[i] <- mean(fr_sample_2010)/mean(fr_sample_2016)
}

quantile(ratiomeans, c(0.025, 0.975))

fr_mean_2010 <- mean(df_france_2010$NO2[df_france_2010$Measurement.Year == '2010'])
fr_mean_2016 <- mean(df_france_2016$NO2[df_france_2016$Measurement.Year == '2016'])
ratio_actual <- fr_mean_2010/fr_mean_2016
bias= round(mean(ratiomeans)-mean(ratio_actual), digits = 3) #estimated bias

print(paste("The estimated bias is", bias))
```

Conclusion: Yes, France has decreased their NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2016.

## Test 2.2: Did India have greater NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2018 than in 2014?  

Additional Welsh t-test is performed to assess if India has greater NO<sub>2</sub> (µg/m<sup>3</sup>) in 2018 compared to 2014. The average NO2 emissions in India in 2018 are also compared to the average NO<sub>2</sub> emissions in India in 2014, as the frequencies for them are almost closer to each other, yet significant enough to assess changes. 

```r
count_df_india <- count_df[count_df$WHO.Country.Name=="India",]
print(count_df_india)
```

Using Welch Two Sample t-test, the null hypothesis and alternative hypothesis are: 

$H_0$: India did not increase NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2018 compared to 2014.  

$H_1$: India increased NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2018 compared to 2014.

```r
df_india <- df[df$WHO.Country.Name=="India", ]

india2014 <- subset(df_india, select=NO2,
                      subset=Measurement.Year==2014, drop=T)

india2018  <- subset(df_india, select=NO2,
                      subset=Measurement.Year==2018, drop=T)

print(t.test(india2018, india2014, alt="greater"))
```

Conclusion: The p-value is greater than 0.05, which fails to reject the null hypothesis.

## Test 3: Was global NO<sub>2</sub> dependent on Year?

To compare if global NO<sub>2</sub> emissions are independent of year, a Chi-square test is performed. In this subset, all countries and cities are included from the cleaned dataset. According to Jarvis et al. (2010), “the annual average indoor nitrogen dioxide guideline of 40 (µg/m<sup>3</sup>), consistent with the existing WHO air quality guideline, is recommended.” The data included in our analysis ranged from 2010 to 2019. Since this guideline is published in the date range of the collected data, NO<sub>2</sub> (µg/m<sup>3</sup>) values that fell at and below 40 (µg/m<sup>3</sup>) annually are categorized in a new column as within the WHO guidelines. NO<sub>2</sub> values that are greater than 40 μg/m<sup>3</sup> are categorized as above.

The new WHO guideline column and the year column meets the required assumption that both variables in the Chi-square test are categorical. Like in t-test, each observation is assumed to be independent. A contingency table is calculated with the number of cities meeting within or above the WHO guidelines for each year. The values in the contingency table are mutually exclusive. 

According to the existing WHO air quality guidelines, a new column is added labeling NO<sub>2</sub> above and within 40 (µg/m<sup>3</sup>) (Jarvis et al., 2010) to the merged dataset. 

```r
df$WHO.Guideline <- ifelse(df$NO2 >= 41.0, "above", "within")
head(df)
```

To assess if global compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) is dependent on year, Chi-squared test is used.  

The null hypothesis and alternative hypothesis are:  

$H_0$: The compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) are independent from years. The compliance with recommended NO<sub>2</sub> does not differ between years.  

$H_1$: The compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) are dependent on the year. The compliance with recommended NO<sub>2</sub> differs between the years.  

```r
df_global <- df[(df$Measurement.Year>=2013 & df$Measurement.Year<=2019),]

# actual
table_year<- table(df_global$Measurement.Year, df_global$WHO.Guideline)
table_year

print(chisq.test(table_year))
```

Additionally, expected values of the contingency table are calculated and compared to the actual values. All expected values of the contingency table are above 50 for all the values. Thus, these criteria confirm the appropriate use of Chi-square test as each value is large enough. 

```r
#expected
print(chisq.test(table_year)$expected)
```


Conclusion: Therefore, compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) differs between the years.

Since the confidence interval includes values greater than 0, the alternative hypothesis is supported. Therefore, France has less NO<sub>2</sub> (µg/m<sup>3</sup>) in 2016 compared to 2010.

To analyze a country within the South East Asia Region, another t-test is performed to see if India had greater emissions in 2014 than in 2014. The results show a p-value greater than 0.05, which fails to reject the null hypothesis. India had greater NO<sub>2</sub> (µg/m<sup>3</sup>) emissions in 2018 compared to 2014.

To determine if NO<sub>2</sub> (µg/m<sup>3</sup>) is dependent on year, Chi-square test is performed. The p-value was much less than 0.05. This result supports the alternative hypothesis. The compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) is dependent on the year. The compliance with recommended NO<sub>2</sub> (µg/m<sup>3</sup>) differs between the years.
